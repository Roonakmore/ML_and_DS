```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/EduData.csv',na_values='?')
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["raisedhands"] = df["raisedhands"].astype(float)
df["raisedhands"].fillna(value = df["raisedhands"].mean() , inplace = True)

df["Topic"].fillna(value = df["Topic"].mode() , inplace = True)

df.dropna(how = 'any' , inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df['Semester'].value_counts()
```

```python
plt.hist(df['Semester'])
plt.title('BAR PLOT')
plt.show()
```

```python
df.drop(df[df['Semester']=='T'].index , inplace=True)
```

```python
plt.hist(df['Semester'])
plt.title('BAR PLOT')
plt.show()
```

```python
df['ParentAnsweringSurvey'].value_counts()
```

```python
df.drop(['ParentAnsweringSurvey'], axis=1,inplace=True)
df
```

```python
df['raisedhands']=df['raisedhands']/ df['raisedhands'].max()
df['VisITedResources']=df['VisITedResources']/ df['VisITedResources'].max()
```

```python
df[['raisedhands','VisITedResources']].head(10)
```
